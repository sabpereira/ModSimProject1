```python
%matplotlib inline

from modsim import *

from pandas import read_html
```

```python
#Importing tables 

filename = 'data/OtterData.html'
tables = read_html(filename, header=1, index_col=1)

table = tables[0]
table.columns = ['years', 'independents', 'pups', 'total']

independents = table.independents
pups = table.pups
total = table.total

total
```

```python
def plot_populations(table):
    """Plot census.
    
    table: DataFrame with columns 'un' and 'census'
    """
    independents = table.independents
    pups = table.pups
    total = table.total
    
    plot(independents, ':', color='darkblue', label='Independents')
    plot(pups, '--', color='green', label='Pups')
    plot(total, color='red', label='Total')
    
    decorate(xlabel='Years',
             ylabel='California Coast Otter Pop')
```

```python
plot_populations(table)
```
